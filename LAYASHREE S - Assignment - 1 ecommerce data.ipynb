### QUESTION- 7 Use the calculation from question - 5 and order priority to check is there any relationship between order priority and days_taken

```python
from scipy.stats import pearsonr
```

```python
# creating the function in order to decode the string values to integers
def recode(value):
    if value == 'Low':
        return 0
    elif value == 'Medium':
        return 1
    elif value == 'High':
        return 2
    else:
        return 3
```

```python
# applying the decode to the variable
merged_df['order_priority_1'] = merged_df['Order Priority'].apply(recode)
merged_df.dtypes
```

```python
# conducting the correlation
r = pearsonr(merged_df['order_priority_1'],merged_df['days_taken'])
r
```

#### as p-value is less than 0.05 therefore there is moderate negative correlation between the two variables, and this relationship is statistically significant
